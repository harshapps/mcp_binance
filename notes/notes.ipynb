````python
# Transport Types in MCP — Notes

## Overview
Transport types define how the MCP (Model Context Protocol) communicates with the LLM. They specify the channel, format, and delivery method for requests and responses between your application and the model.

## Common Transport Types
- HTTP/HTTPS  
    - Standard web requests (REST-like).  
    - Easy to debug and integrate with web services.
- WebSocket  
    - Persistent, bidirectional connection.  
    - Good for streaming tokens or real-time interactions.
- gRPC  
    - Binary protocol with strong typing and performance.  
    - Useful for high-throughput, low-latency systems.
- Local IPC (Unix socket / Named pipe)  
    - Fast local communication between processes.  
    - Suitable when LLM runs on the same host.
- File-based / Queue (e.g., message queues)  
    - Indirect transport via files or queues (RabbitMQ, Kafka).  
    - Useful for asynchronous or decoupled workflows.

## How They Work (Simple)
- Request sent by MCP → Transport serializes the request → LLM receives and processes → Response serialized back → MCP deserializes and uses the result.
- Streaming transports (WebSocket, gRPC streaming) deliver partial results as they are produced.

## When to Use Which
- Rapid development & public APIs: HTTP/HTTPS.
- Real-time or streaming use cases: WebSocket or gRPC streaming.
- High-performance internal services: gRPC or IPC.
- Asynchronous / decoupled systems: Message queues or file-based transports.

## Pros & Cons (High-level)
- HTTP: simple, compatible, higher overhead for many small messages.
- WebSocket: low-latency streaming, more complex connection management.
- gRPC: efficient and typed, requires protobufs and extra setup.
- IPC: very fast locally, not networked.
- Queues: resilient and decoupled, adds latency and complexity.

## Best Practices
- Match transport to latency and throughput needs.
- Use streaming transports when you need partial outputs or lower perceived latency.
- Secure transports (TLS, authentication) for networked communications.
- Add retries, timeouts, and backpressure handling for robustness.
- Monitor metrics per transport (latency, errors, connection health).

## Troubleshooting Tips
- If latency is high, check serialization overhead and network hops.
- For dropped streaming messages, verify connection keep-alive and buffer sizes.
- For mismatched schema errors, validate message formats and versions (e.g., protobuf definitions).
- Inspect raw requests/responses (logs or network traces) to debug payload issues.

## Quick Summary
Choose the transport based on real-time needs, performance, and deployment topology. HTTP for simplicity, WebSocket/gRPC for streaming and low latency, IPC for fast local calls, and queues for asynchronous resilience.
````
